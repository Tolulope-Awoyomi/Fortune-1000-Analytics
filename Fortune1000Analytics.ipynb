```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming your dataset is loaded into a pandas DataFrame called 'df'
# Replace 'your_data.csv' with the actual file name or path if you're loading from a CSV file
# df = pd.read_csv('your_data.csv')

# Filter out rows with missing rank_change values
filtered_df = dataset.dropna(subset=['rank_change'])

# Sort the DataFrame by rank_change in ascending order
sorted_by_rank_change = filtered_df.sort_values(by='rank_change', ascending=True)

# Get the top 10 companies with the largest positive rank changes (top movers)
top_decliners = sorted_by_rank_change.head(10)

# Get the top 10 companies with the largest negative rank changes (top decliners)
top_movers = sorted_by_rank_change.tail(10)

# Create subplots for top movers and top decliners
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Bar plot for top movers
ax1.barh(top_movers['company'], top_movers['rank_change'], color='green')
ax1.set_xlabel('Rank Change from previous year (2021)')
ax1.set_ylabel('Company')
ax1.set_title('Top Movers')

# Bar plot for top decliners
ax2.barh(top_decliners['company'], top_decliners['rank_change'], color='red')
ax2.set_xlabel('Rank Change from previous year (2021)')
ax2.set_ylabel('Company')
ax2.set_title('Top Decliners')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming your dataset is loaded into a pandas DataFrame called 'df'
# Replace 'your_data.csv' with the actual file name or path if you're loading from a CSV file
# df = pd.read_csv('your_data.csv')

# Filter out rows with missing rank_change values
filtered_df = dataset.dropna(subset=['rank_change'])

# Select the columns for correlation analysis
correlation_columns = ['rank_change', 'revenue', 'profit', 'num. of employees']

# Create a subset DataFrame with selected columns
correlation_df = filtered_df[correlation_columns]

# Calculate the correlation matrix
correlation_matrix = correlation_df.corr()

# Create a heatmap to visualize the correlations
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap')
plt.tight_layout()

# Show the plot
plt.show()
```
